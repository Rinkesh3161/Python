```python
user_data = pd.read_csv("https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv")
```

```python
user_data.head()
```

```python
session_data = pd.read_csv("https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv")
session_data.head()
```

```python
transaction_data = pd.read_csv("https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv")
transaction_data.head()
```

```python
merge_result =transaction_data.merge(user_data, on='UserID', how='left')
merge_result
```

```python
# merge_result['User'== "NaN"]
# #transactions[transactions.UserID == 2]
transaction_data[transaction_data['UserID'].isin(user_data['UserID'])]
```

```python
transaction_data[~transaction_data['UserID'].isin(user_data['UserID'])]
```

```python
merge_result1 =transaction_data.merge(user_data, on='UserID', how='inner')
merge_result1
```

```python
merge_result2 =transaction_data.merge(user_data, on='UserID', how='outer')
merge_result2
```

```python
#Determine which sessions occurred on the same day each user registered

pd.merge(left=user_data, right=session_data, how='inner', left_on=['UserID', 'Registered'], right_on=['UserID', 'SessionDate'])
```

```python
pd.merge(transaction_data, transaction_data, on='UserID')
```

```python
# Join each user to his/her first occuring transaction in the transactions table
pd.merge(user_data, transaction_data.groupby('UserID').first().reset_index(), how='left', on='UserID')
```
